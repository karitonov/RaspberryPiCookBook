# 目次の読み込み

```python
with open("Contents.txt", "r", encoding="utf-8") as f:
    lines = f.readlines()

# print(lines)

with open("Contents.md", "w", encoding="utf-8") as f:
    chap = 0
    for line in lines:
        if len(line) > 1:
            if line[0].isalnum():
                chap += 1
                sec = 0
                new_line = "# " + line.lstrip()
            elif line[4].isalnum():
                sec += 1
                pla = 0
                new_line = f"## {line.lstrip()}"
            elif line[8].isalnum():
                pla += 1
                new_line = f"### {chap}.{sec}.{pla} {line.lstrip()}"
        else:
            new_line = line

        print(new_line, end="")
        f.writelines(new_line)
```
